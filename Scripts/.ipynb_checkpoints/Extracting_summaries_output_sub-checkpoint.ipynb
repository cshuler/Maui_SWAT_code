```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Datapath of raw unzipped files
datapath= os.path.join("..", 'Raw_Data_Hla', 'Unzipped')

#datapath= os.path.join("..", 'Raw_Data_Hla', 'Example_J2017_EastWestMaui', 'J2017_EastWestMaui')


# file_path = os.path.join(datapath, "Scenarios", 'Default', 'TxtInOut')
```

```python
os.listdir(datapath)
```

```python
for dir in os.listdir(datapath):
    print('\n \n {}'.format(dir))
    go = os.path.join(datapath, dir)
    
    dir_a = next(os.walk(go))[1]
    print(f"**{dir_a}")
    
    for a in dir_a:
      go = os.path.join(datapath, dir, a)
      dir_b = next(os.walk(go))[1]
      print(f"....{dir_b}")

      for b in dir_b:
        print(b)
        go = os.path.join(datapath, dir, a, b)
        dir_c = next(os.walk(go))[1]
        print(f"------------{dir_c}")

        for c in dir_c:
          print(c)
          go = os.path.join(datapath, dir, a, b, c)
          dir_d = next(os.walk(go))[1]
          print(f"_-_-_-_-__-_-_-_-{dir_d}")
```

```python
def list_files(startpath):
    for root, dirs, files in os.walk(startpath):
        level = root.replace(startpath, '').count(os.sep)
        indent = ' ' * 4 * (level)
        print('{}{}/'.format(indent, os.path.basename(root)))
        subindent = ' ' * 4 * (level + 1)
        for f in files:
            print('{}{}'.format(subindent, f))
```
